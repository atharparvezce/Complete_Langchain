---

**LangChain** is a framework that helps you build applications using **large language models (LLMs)** by integrating them with external tools, data sources, and APIs.

---

# Key Features:
- Connects LLMs to external data (e.g., documents, APIs).
- Supports multi-step workflows (chains).
- Enables memory, tool usage, and agent-based applications.

---

# What are LangChain Models

LangChain simplifies using different types of AI models:

1. **Language Models**
    - LLMs (Text completion)
    - Chat Models (Conversational agents)
2. **Embedding Models**
    - Convert text into vectors for similarity and retrieval

---

# 1. Language Models

## a) LLMs (Text Completion Models)

These models take a text prompt and return a continuation.

**Examples:**
- OpenAI `text-davinci-003`
- Claude
- Google PaLM

## b) Chat Models
Optimized for conversational use, working with message objects instead of plain text.

### Closed-source:

OpenAI ChatGPT (gpt-3.5-turbo, gpt-4)
Anthropic Claude

### Open-source:

Meta LLaMA 2
Mistral, Zephyr

---

# 2. Embedding Models

These models convert text into vector representations (useful for similarity search, document retrieval, etc.).

## Closed-source:
OpenAI Embeddings (text-embedding-ada-002)
Cohere

## Open-source:
HuggingFace all-MiniLM-L6-v2
Sentence Transformers
